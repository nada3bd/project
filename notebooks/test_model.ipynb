```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

model = YOLO('/content/detect_doctor_patient.pt')

# 2. تشغيل الموديل على صورة الاختبار
results = model.predict(source='/content/caring-for-patients-and-providing-thumbnail-173396.webp', conf=0.5)

# 3. معالجة وعرض النتائج
for r in results:
    im_array = r.plot() # رسم المربعات والأسماء

    # تحويل الألوان من BGR إلى RGB لكي تظهر الألوان بشكل صحيح
    im_rgb = cv2.cvtColor(im_array, cv2.COLOR_BGR2RGB)

    # استخدام matplotlib للعرض بدل cv2_imshow
    plt.figure(figsize=(10, 10))
    plt.imshow(im_rgb)
    plt.axis('off') # إخفاء محاور الرسم
    plt.show()
```
